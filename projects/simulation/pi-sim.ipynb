```python
import plotly.express as px
import numpy as np
from IPython.display import display, clear_output
import time

# Number of darts to throw
num_darts = 10000

# Generate random x and y coordinates
x = np.random.uniform(-1, 1, num_darts)
y = np.random.uniform(-1, 1, num_darts)



# Create the scatter plot
import plotly.graph_objects as go

fig = go.Figure()
N = 100
NUM_BATCHES = num_darts // N
total_inside_circle = 0
x1 = []
y1 = []
# Create the scatter plot
for i in range(0, NUM_BATCHES):
    arr_size = (i+1)*N
    # x1=x[:arr_size]
    # y1=y[:arr_size]
    rnd = np.round(np.random.uniform(-1, 1, N), 3)
    x1.extend(rnd)
    rnd = np.round(np.random.uniform(-1, 1, N), 3)
    y1.extend(rnd)
    # Calculate distance from the origin
    distance = np.sqrt(np.pow(x1,2) + np.pow(y1,2))
    # Determine if the dart is inside the circle
    inside_circle = distance <= 1

    total_inside_circle += np.sum(inside_circle)

    # Calculate the value of pi
    pi_estimate = 4 * total_inside_circle / arr_size

    fig = go.Figure(data=go.Scatter(
        x=x1,
        y=y1,
        mode='markers', 
        marker=dict(color=['red' if inside else 'blue' for inside in inside_circle])
    ))
    fig.update_layout(
        title=f'<b>Estimate of π using Monte Carlo Simulation<b> <br>Num Darts: {arr_size}<br>Estimated π: {pi_estimate:.3f}',
        title_x=0.5,
        xaxis_title='x',
        yaxis_title='y',
        height=600,
        width=600,
        showlegend=False,
        title_font=dict(size=20, family='Courier New', color='black')
    )
    # Add a circle for reference
    circle_x = np.linspace(-1, 1, 400)
    circle_y = np.sqrt(1 - circle_x**2)
    fig.add_scatter(x=circle_x, y=circle_y, mode='lines', name='Circle', line=dict(color='red'))
    fig.add_scatter(x=circle_x, y=-circle_y, mode='lines', name='Circle', line=dict(color='red'))
    # Add a square for reference
    square_x = [-1, 1, 1, -1, -1]
    square_y = [-1, -1, 1, 1, -1]
    fig.add_scatter(x=square_x, y=square_y, mode='lines', name='Square', line=dict(color='blue'))
    clear_output(wait=True)
    display(fig)
    time.sleep(0.1)
```
